```python
#Calculando o lucro de cada item
data["Lucro"]= (data['receita_bruta']*data['margem']/100).round(2)
data.head()
```

```python
data2 = data[['item_descricao','item_unidade', 'Lucro']].groupby(['item_descricao']).sum(['Lucro'])
data2.head()
```

```python
#Os 10 itens de maior relevância no período
data2.sort_values(by='Lucro',ascending= False).head(10)
```

```python
periodo = data['periodo'].unique()
periodo
```

```python
#os top 15 itens que acompanham cada um dos período.
data3 = data[['periodo','item_descricao','item_unidade', 'Lucro']].groupby(['periodo','item_descricao']).sum(['Lucro'])
data3=data3.reset_index()

for x in periodo:
    print(data3.loc[data3['periodo']==x].sort_values(by='Lucro',ascending= False).head(15))


```

```python
#Criando/Analisando a possibilidade de combos promocionais...
data4= data.drop(data[data['item_descricao']=='SACOLA PLASTICA MM BRANCA'].index)
data4
```

```python
data4['Agrupado'] = data4.groupby(data4['socio'])['item_descricao'].transform(lambda x: '---'.join(x))
```

```python
data5 = data4[['socio','Agrupado']].drop_duplicates()
data5
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()
```

```python
#Encontrando possíveis combos...
for row in data5['Agrupado']:
    row_list = row.split('---')
    count.update(Counter(combinations(row_list,2)))

combos=[]
for key, value in count.most_common(10):
    print(key,value)
    combos.append(key)
```

```python
print("Possíveis combos:", combos)
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
data_te = te.fit(combos).transform(combos)
```

```python
df = pd.DataFrame(data_te,columns = te.columns_)
df.head()
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
combos2 = apriori(df, min_support = 0.003, use_colnames= True)
combos2.head()
```

```python
combos2.sort_values(by=['support'], ascending = False)
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
rules = association_rules(combos2, metric = 'confidence', min_threshold = 0.001)
rules.columns
```

```python
analise = rules.sort_values(by= ['lift'], ascending = False).drop(['antecedent support',
       'consequent support', 'support', 'confidence', 'lift', 'leverage',
       'conviction', 'zhangs_metric'], axis=1)

analise
```

```python
#Encontrando Novos Combos
antecedente = [list(x) for x in analise['antecedents']]
lista = sum(antecedente, [])

consequente = [list(x) for x in analise['consequents']]
lista2 = sum(consequente, [])


for itemA in lista:
    for i, tupla in enumerate(combos):
        if itemA in tupla:
            itemB = lista2[lista.index(itemA)]
            if itemB not in tupla:
                nova_tupla = tupla + (itemB,)
                if len(nova_tupla) == 3:
                    combos[i] = nova_tupla
                    break

print("Novos combos: ", combos)
```
